```
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import os

path = kagglehub.dataset_download("mohansacharya/graduate-admissions")
print("Path to dataset files:", path)

df = pd.read_csv(path + '/Admission_Predict.csv')
print("–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")

#–ü–æ–∏—Å–∫ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
print("\n–°—Ç–æ–ª–±—Ü—ã –¥–∞—Ç–∞—Å–µ—Ç–∞:")
print(df.columns.tolist())

target_keywords = ['chance', 'admit', 'probability', 'rate', 'score']
target = None

for col in df.columns:
    if any(keyword in col.lower() for keyword in target_keywords):
        target = col
        break


print(f"\n –¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫: {target}")

print(f"\n –ê–Ω–∞–ª–∏–∑ —Ñ–∞–∫—Ç–æ—Ä–æ–≤, –≤–ª–∏—è—é—â–∏—Ö –Ω–∞ {target}:")

numeric_df = df.select_dtypes(include='number')

if target in numeric_df.columns:
    correlations = numeric_df.corr()[target].sort_values(ascending=False)

    print("\n–§–∞–∫—Ç–æ—Ä—ã, –≤–ª–∏—è—é—â–∏–µ –Ω–∞ –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏–µ:")
    for factor, corr in correlations.items():
        if factor != target:
            importance = "üî¥ –í—ã—Å–æ–∫–∞—è" if abs(corr) > 0.5 else "üü° –°—Ä–µ–¥–Ω—è—è" if abs(corr) > 0.3 else "üü¢ –ù–∏–∑–∫–∞—è"
            print(f"{importance} {factor}: {corr:.3f}")
```
